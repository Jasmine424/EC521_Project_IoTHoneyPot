```python
import ijson
import io

full = []
filename = "cowrie4.json"
with open(filename, 'r') as f:
    cursor = 0
    for line_number, line in enumerate(f):
        print ("Processing line", line_number + 1,"at cursor index:", cursor)
        line_as_file = io.StringIO(line)
        # Use a new parser for each line
        json_parser = ijson.parse(line_as_file)
        for prefix, type, value in json_parser:
            if prefix == 'src_ip':
                #print ("prefix=",prefix, "type=",type, "value=",value)
                full.append(value)
        cursor += len(line)

#df = pd.read_json('~/Downloads/cowrie4.json', lines=True)['src_ip'].unique() #parse and get unique addresses into list
#print(df)
```

```python
import requests
import re

df = set(full)
print(df)
count = 0
human_count23 = 0
human_count22 = 0
for i in df:
    m = 'https://www.shodan.io/host/' + i + ''
    info = (requests.get(m).text)
    soup = BeautifulSoup(info, 'html.parser')
    for tag in soup.find_all("meta"):
        if tag.get("name", None) == "twitter:description":
            ports = (tag.get("content", None))
            count += 1
            print(i, ports)
            if re.search("22", ports):
                human_count22 += 1
            if re.search("23", ports):
                human_count23 += 1 
print("total ip:", count)
print("ssh open {0}%".format(human_count22/int(count) * 100))
print("telnet open {0}%:".format(human_count23/int(count)* 100))

    
```
