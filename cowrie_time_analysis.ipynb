```python
import requests
import re

df = set(full)
print(df)
count = 0
human_count23 = 0
human_count22 = 0
for i in df:
    m = 'https://www.shodan.io/host/' + i + ''
    info = (requests.get(m).text)
    soup = BeautifulSoup(info, 'html.parser')
    for tag in soup.find_all("meta"):
        if tag.get("name", None) == "twitter:description":
            ports = (tag.get("content", None))
            count += 1
            print(i, ports)
            if re.search("22", ports):
                human_count22 += 1
            if re.search("23", ports):
                human_count23 += 1 
print("total ip:", count)
print("ssh open {0}%".format(human_count22/int(count) * 100))
print("telnet open {0}%:".format(human_count23/int(count)* 100))

    
```
